```python
def drop_nan_by_count_column(data, count):
    nan_counts = data.isna().sum()

    # Get the column names where the count of NaN values is greater than 5
    columns_to_drop = nan_counts[nan_counts > count].index

    # Drop the columns from the DataFrame
    df = data.drop(columns_to_drop, axis=1)
    return df
```
