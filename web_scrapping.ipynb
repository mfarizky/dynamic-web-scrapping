```python
import requests
import json
import pandas as pd
import datetime
```

```python
def generate_custom_dates(start_date, end_date):
    date_format = "%Y-%m-%d"
    start = datetime.datetime.strptime(start_date, date_format)
    end = datetime.datetime.strptime(end_date, date_format)
    dates = []

    current_date = start
    while current_date <= end:
        dates.append(current_date.strftime(date_format))
        current_date += datetime.timedelta(days=1)

    return dates

def get_json(date):
    #print(date)
    URL= 'https://siskaperbapo.jatimprov.go.id/home2/getDataMap/?tanggal='+ date +'&komoditas=62'
    #print(URL)
    response = requests.get(URL, 
                    headers={'Content-Type': 'text/html; charset=UTF-8'
                            },
                    data={'tanggal':date, 'komoditas':'62'})
    json_data = json.loads(response.text)
    data = json_data
    return data

def get_data_fromjson(data):
    # data_list = []
    # Mendapatkan tanggal
    tanggal = data['tanggal']

    # Membangun data CSV dalam bentuk kamus
    csv_data = {'tanggal': tanggal}
    for key, value in data['data'].items():
        if key != 'tanggal' and value['hrg'] != 0:
            csv_data[f'harga_{value["code"]}'] = value['hrg']
    return csv_data

    # Mengonversi data ke dalam dataframe pandas
  
    
    
    
    # combined_data = {'tanggal': [], 'harga': []}
    # for data in data_list:
    #     tanggal = data['tanggal']
    #     for key, value in data['data'].items():
    #         if key != 'tanggal' and value['hrg'] != 0:
    #             combined_data['tanggal'].append(tanggal)
    #             combined_data['harga'].append(value['hrg'])

    # # Mengonversi data ke dalam dataframe pandas
    # df = pd.DataFrame(combined_data)
    # df.to_csv('data.csv', index=False)

    # def get_data(data):
    # # Mendapatkan tanggal
    #     tanggal = data['tanggal']

    #     # Membangun data CSV dalam bentuk kamus
    #     csv_data = {'tanggal': tanggal}
    #     for key, value in data['data'].items():
    #         if key != 'tanggal' and value['hrg'] != 0:
    #             csv_data[f'harga_{value["code"]}'] = value['hrg']

    #     # Mengonversi data ke dalam dataframe pandas
    #     df = pd.DataFrame([csv_data])
    #     df.to_csv('data.csv', index=False)
```

```python
# Custom Tanggalmu Disini
start_date = "2023-05-14"
end_date = "2023-05-24"
custom_dates = generate_custom_dates(start_date, end_date)
```

```python
data_list = []
for date in custom_dates:
    print(date)
    link = get_json(date)
    # print (link)
    data = get_data_fromjson(link)
    data_list.append(data)
```

```python
df = pd.DataFrame(data_list)
df.to_csv('File_test.csv', index=False)
```
